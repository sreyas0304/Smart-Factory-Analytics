```scala
%flink.ssql
DROP TABLE IF EXISTS sensor_source;

CREATE TABLE sensor_source (
    event_id STRING,
    device_id STRING,
    device_type STRING,
    zone STRING,
    event_time STRING,
    temperature_c DOUBLE,
    vibration_x DOUBLE,
    motor_current_amps DOUBLE,
    
    -- THE FIX: If event_time is missing/bad, default to 1970-01-01
    window_time AS COALESCE(
        TO_TIMESTAMP(SUBSTRING(event_time, 1, 23), 'yyyy-MM-dd''T''HH:mm:ss.SSS'),
        TO_TIMESTAMP('1970-01-01 00:00:00', 'yyyy-MM-dd HH:mm:ss')
    ),
    
    WATERMARK FOR window_time AS window_time - INTERVAL '10' SECOND
) WITH (
    'connector' = 'kinesis',
    'stream' = 'smart-factory-telemetry',
    'aws.region' = 'us-east-1',
    'format' = 'json',
    'json.ignore-parse-errors' = 'true',
    'scan.stream.initpos' = 'LATEST'
);
```

```scala
%flink.ssql
-- 1. Clear the existing view from the catalog
DROP VIEW IF EXISTS live_dashboard_metrics;

-- 2. Create the updated view
CREATE VIEW live_dashboard_metrics AS
SELECT 
    device_id,
    device_type,
    zone,
    TUMBLE_END(window_time, INTERVAL '1' MINUTE) as window_end_time,
    ROUND(AVG(temperature_c), 2) as avg_temp,
    ROUND(MAX(vibration_x), 2) as peak_vibration
FROM sensor_source
GROUP BY 
    device_id,
    device_type,
    zone,
    TUMBLE(window_time, INTERVAL '1' MINUTE);
```

```scala
%flink.ssql
-- Step 3: 5-Minute Moving Average (Updated every 1 min)

-- 1. Clear the existing view from the catalog for safe re-runs
DROP VIEW IF EXISTS sustained_load_metrics;

-- 2. Create the updated view using the watermarked column
CREATE VIEW sustained_load_metrics AS
SELECT 
    device_id,
    HOP_END(window_time, INTERVAL '1' MINUTE, INTERVAL '5' MINUTE) as window_end_time,
    ROUND(AVG(motor_current_amps), 2) as moving_avg_current
FROM sensor_source
WHERE device_type = 'conveyor'
GROUP BY 
    device_id,
    HOP(window_time, INTERVAL '1' MINUTE, INTERVAL '5' MINUTE);
```

```scala
%flink.ssql
-- Step 4: Define DynamoDB Sink

-- 1. Clear the existing table blueprint from the catalog for safe re-runs
DROP TABLE IF EXISTS dynamodb_machine_state;

-- 2. Create the updated sink table definition
CREATE TABLE dynamodb_machine_state (
    device_id STRING,
    device_type STRING,
    zone STRING,
    last_updated TIMESTAMP(3),
    avg_temp DOUBLE,
    peak_vibration DOUBLE,
    PRIMARY KEY (device_id) NOT ENFORCED
) WITH (
    'connector' = 'dynamodb',
    'table-name' = 'FactoryMachineState',
    'aws.region' = 'us-east-1'
);

-- 3. Continuously stream the Tumbling Window results to DynamoDB
INSERT INTO dynamodb_machine_state
SELECT 
    device_id,
    device_type,
    zone,
    window_end_time as last_updated,
    avg_temp,
    peak_vibration
FROM live_dashboard_metrics;
```
